# Eigenfaces

```python
import numpy as np
import sklearn.decomposition
from sklearn.datasets import fetch_olivetti_faces
import matplotlib.pyplot as plt

plt.figure(figsize=(8,8))

%matplotlib inline
```

Let's download a dataset of face images.

```python
dataset = fetch_olivetti_faces(shuffle=True)
faces = dataset.data
```

```python
plt.imshow(faces[5].reshape(64,-1))
```

We'll perform PCA on this dataset.

```python
pca = sklearn.decomposition.PCA(n_components=10)
pca.fit(faces)
```

What do the eigenvectors of $X X^\intercal$ look like? We can visualize them as images:

```python
for i in range(10):
    plt.figure()
    plt.title(f'Eigenvector {i+1}')
    plt.imshow(pca.components_[i].reshape(64,-1))
```

We see that some of the eigenvectors are "detectors". For instance, we might say that eigenvector 7 is a "eyegalss detector". If an image contains eyeglasses, the norm of its projection onto this eigenvector will be large.
